```python
def count_many(needles, haystack):
    count = Counter(haystack.split())
    return {key: count[key] for key in count if key in needles}
```

```python
def get_all_counts(dictio):
    val = 0
    for key in dictio:
        val += dictio[key]
    return val
```

```python
def count_words(text):
    counter = 0
    text_split = text.split()
    for word in text_split:
        if (len(word) > 5):
            counter+=1
    return counter
```

```python
def month_converter(month):
    months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
    return months.index(month) + 1
```

```python
def getData(url):
    photos = 0
    n = 13
    X = np.empty(shape=[0, n])
    datepast = datetime.datetime.now()
    unique_tags = []
    
    for i in range(1,16):
        feed = ""
        if i == 1:
            feed = feedparser.parse(url+'/rss')
        else:
            print(str(i))
            feed = feedparser.parse(url+'/page/'+str(i)+'/rss')
        for entry in feed['entries']:
            if 'summary' in entry:
                #print(entry)

                #date published
                date = entry['published'] 
                #get information, date - time - title - text - tags
                info = date.split()

                time = info[4].split(":")
                time_elapse = (datepast-datetime.datetime(int(info[3]),month_converter(info[2]),int(info[1]),int(time[0]),int(time[1]),int(time[2]))).total_seconds()/60 #datetime(year, month, day
                datepast = (datetime.datetime(int(info[3]),month_converter(info[2]),int(info[1]),int(time[0]),int(time[1]),int(time[2])))
                tag_self = 0

                if 'tags' in entry:
                    for term in entry['tags']:
                        tag_self += 1
                        if term not in unique_tags:
                            unique_tags.append(term)
                self_words = get_all_counts(count_many(['me','my','mine','I','myself','I\'m'], BeautifulSoup(entry['summary']).text))
                articles = get_all_counts(count_many(['a', 'an', 'the'], BeautifulSoup(entry['summary']).text))
                themes = get_all_counts(count_many(['family', 'friend', 'friends', 'work', 'office', 'home', 'house'], BeautifulSoup(entry['summary']).text))
                summary_words = count_words(BeautifulSoup(entry['summary']).text)               
                has_image = 0
                photos = entry['summary'].count("<img")
                if photos>=1:
                    has_image = 1
                text = TextBlob(BeautifulSoup(entry['summary']).text)
                polarity = text.sentiment.polarity
                subjectivity = text.sentiment.subjectivity
                sentences = len(text.sentences)
                words = len(text.words)
                X = np.append(X, [[time_elapse, self_words, articles, themes, tag_self, summary_words, has_image,
                                   polarity, subjectivity, sentences, words, photos, len(unique_tags)]], axis=0)
    return X
```
