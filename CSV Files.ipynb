```python
def count_many(needles, haystack):
    count = Counter(haystack.split())
    return {key: count[key] for key in count if key in needles}
```

```python
def get_all_counts(dictio):
    val = 0
    for key in dictio:
        val += dictio[key]
    return val
```

```python
def count_words(text):
    counter = 0
    for word in text.split(r'[^\w]'):
        if (len(word) > 5):
            counter+=1
    return counter
```

```python
def month_converter(month):
    months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
    return months.index(month) + 1
```

```python
def getData(url):
    photos = 0
    n = 12
    X = np.empty(shape=[0, n])

    for i in range(1,16):
        try:
            feed = ""
            if i == 1:
                feed = feedparser.parse(url+'/rss')
            else:
                print(str(i))
                feed = feedparser.parse(url+'/page/'+str(i)+'/rss')
            for entry in feed['entries']:
                if 'summary' in entry:
                    #print(entry)

                    #date published
                    date = entry['published'] 
                    #get information, date - time - title - text - tags
                    info = date.split()

                    time = info[4].split(":")
                    time_elapse = (datepast-(datetime.datetime(int(info[3]),month_converter(info[2]),int(info[1])))).total_seconds()/60 #datetime(year, month, day

                    tag_self = 0

                    if 'tags' in entry:
                        for term in entry['tags']:
                            tag_self += 1
                            if term not in tags:
                                tags[(term)] = 1
                            else:
                                tags[(term)]+=1
                    self_words = get_all_counts(count_many(['me','my','mine','I','myself','I\'m'], BeautifulSoup(entry['summary']).text))
                    articles = get_all_counts(count_many(['a', 'an', 'the'], BeautifulSoup(entry['summary']).text))
                    themes = get_all_counts(count_many(['family', 'friend', 'friends', 'work', 'office', 'home', 'house'], BeautifulSoup(entry['summary']).text))
                    tags_all = get_all_counts(tags)
                    summary_words = count_words(BeautifulSoup(entry['summary']).text)
                    has_image = 0
                    if "<img" in entry['summary']:
                        photos += 1
                        has_image = 1
                    text = TextBlob(BeautifulSoup(entry['summary']).text)
                    polarity = text.sentiment.polarity
                    subjectivity = text.sentiment.subjectivity
                    sentences = len(text.sentences)
                    words = len(text.words)
                    X = np.append(X, [[time_elapse, self_words, articles, themes, tags_all, summary_words, has_image, polarity, subjectivity, sentences, words, len(tags)]], axis=0)
        except:
            pass
    return X
```

```python
def returnAll(array_all):
    res = []
    for i in range(0, len(array_all)):
        x = []
        for j in range(0,12):
            x.append(np.mean(array_all[i][j,j]))
        res.append(x)
    return np.array(res)
```
